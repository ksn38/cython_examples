```python
def plot1(i, ii):
    jl = mb = mandelbrot_image_circle.calc_mand(1440, 50, 2.5, -0.5, 0, i/100)

    fig, ax = plt.subplots()
    fig.set_size_inches(20, 20)
    fig.patch.set_visible(False)
    ax.axis('off')
    fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
        
    ax.imshow(jl.imag, cmap='tab20b')

    plt.savefig('C:\\Users\\ksn\\temp/{:04d}.png'.format(ii))
    plt.clf()
```

```python
plot1(200, 2)
```

```python
counter = 0
ii = 0

for i in range(1, 100000):
    if np.log10(i) - counter < 2.5:
        counter = 0
        plot1(i, ii)
        #print(i)
        if i == 200:
            k = 0
            while k < 45:
                ii += 1
                plot1(i, ii) 
                k += 1
        ii += 1
    else:
        try:
            counter += 1/((i - ii)**(0.4))
        except ZeroDivisionError:
            pass
```
