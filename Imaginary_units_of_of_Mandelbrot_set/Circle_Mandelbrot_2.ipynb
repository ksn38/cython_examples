```python
counter = 0
ii = 0
x = []
y = []
board = 2.829
exp = 0.6

for i in range(21, 100000): #4000000
    if np.log10(i) - counter < board:
        counter = 0
        ii += 1
        x.append(i)
        y.append(ii)
    else:
        counter += 1/(i**exp)
        
print(ii)
plt.scatter(x, y, s=1)
```

```python
malevich_cmap = matplotlib.colors.ListedColormap(['#1c1b41', '#edcf02', '#38703f', '#f5f7db', \
                '#292328', '#e42417', '#546da2', '#d8b21e'], name='malevich_cmap')

kandinsky_cmap = matplotlib.colors.ListedColormap(['#417145', '#0f1f60', '#b5002f', '#171014', '#f1d625', '#b4d2e3', \
                                                   '#417145', '#331d90', '#b5002f', '#171014', '#f1d625', '#b4d2e3', \
                                                   '#f5e119', '#6d93ab', '#171014', '#fa8c3b', '#0f1f60', '#106848', \
                                                  '#f5e119', '#6d93ab', '#171014', '#fa8c3b', '#6071b4', '#106848',], name='kandinsky_cmap')

picasso_cmap = matplotlib.colors.ListedColormap(['#4d1a08', '#ead319', '#345498', '#a23e32', '#0b0905', '#9eb3d4', \
                                                '#e3e3d4', '#d0cfc8', '#191820', '#c11305', '#b7c908', '#203286',], name='picasso_cmap')
```

```python
def plot1(i, ii):
    md = mandelbrot_image_circle.calc_mand(720, 50, 2.5 + i/400, -0.5, 0, i/100)

    fig, ax = plt.subplots()
    fig.set_size_inches(10, 10)
    fig.patch.set_visible(False)
    ax.axis('off')
    fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
    
    ax.imshow(np.rot90(md.imag, 2), cmap=picasso_cmap)

    plt.savefig('E:\\Temp\\tmp/{:04d}.png'.format(ii))
    plt.clf()
```
