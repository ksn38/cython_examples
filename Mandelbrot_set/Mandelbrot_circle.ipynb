```python
import mandelbrot2

t1 = time.time()
#mandelbrot2.calc_mand(resolution, precision, scale, x, y, exponent)
mb = mandelbrot2.calc_mand(1440, 50, 5, -0.5, 0, 1.7)
print ("time:", time.time() - t1)

plt.subplots()[0].set_size_inches(7, 7)
plt.imshow((mb), cmap='magma')
```

```python
#exponent changes
def plot1(i, ii):
    jl = mb = mandelbrot2.calc_mand(1440, 50, 5, -0.5, 0, i/100)

    fig, ax = plt.subplots()
    fig.set_size_inches(20, 20)
    fig.patch.set_visible(False)
    ax.axis('off')
    fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
        
    ax.imshow(jl, cmap='magma')

    plt.savefig('C:\\Users\\ksn\\temp/{:04d}.png'.format(ii))
    plt.clf()
```

```python
counter = 0
ii = 0

for i in range(1, 100000):
    if np.log10(i) - counter < 2.5:
        counter = 0
        plot1(i, ii)
        #print(i)
        if i == 200:
            k = 0
            while k < 45:
                ii += 1
                plot1(i, ii) 
                k += 1
        ii += 1
    else:
        try:
            counter += 1/((i - ii)**(0.4))
        except ZeroDivisionError:
            pass
```
