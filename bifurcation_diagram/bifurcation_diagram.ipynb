```python
import numpy as np
from matplotlib import pyplot as plt


max_iterations = 100 #max_iterations = 11 for Bifurcation diagram #2

x = 400

p_m = np.zeros((x, max_iterations),dtype=np.complex128)
r_m = np.zeros((x, max_iterations),dtype=np.complex128)


for r in range(0, x):
    m = 1
    population = 1/100
    k = r/100
    while m < max_iterations:
        population = k*(population - population**2)
        #print(r, population)
        p_m[r, m] = population
        r_m[r, m] = k
        m += 1

color = 'white'
fig, ax = plt.subplots(facecolor=color)
ax.set_facecolor(color)
fig.set_size_inches(60, 30)
ax.set_ylim([0, 1])
ax.set_xlim([2, 4])
ax.yaxis.label.set_color(color)
ax.xaxis.label.set_color(color)
ax.tick_params(labelcolor=color)
red = 0.0
blue = 1.0
incr = 1/(max_iterations - 1)
for i in range(max_iterations):
#         print(red, blue)
    ax.scatter(r_m[:, i], p_m[:, i], s=100, c=[[red, 0.0, blue]])
    red += incr
    blue -= incr
plt.show()
#     p_m = np.zeros((20, max_iterations))
#     r_m = np.zeros((20, max_iterations))
```
